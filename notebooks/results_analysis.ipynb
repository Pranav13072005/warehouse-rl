# Results Analysis

This notebook reads benchmark outputs and plots a quick comparison.

Expected artifact: `runs/benchmark.csv` (from `python -m evaluate.benchmark ...`).

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

benchmark_path = Path('..') / 'runs' / 'benchmark.csv'
print('Looking for:', benchmark_path.resolve())

if not benchmark_path.exists():
    raise FileNotFoundError('Run `python -m evaluate.benchmark ...` to create runs/benchmark.csv')

df = pd.read_csv(benchmark_path)
df
```

```python
df_plot = df.sort_values('avg_reward', ascending=True)
plt.figure(figsize=(6, 3))
plt.barh(df_plot['agent'], df_plot['avg_reward'])
plt.title('Average Episode Reward (higher is better)')
plt.xlabel('avg_reward')
plt.tight_layout()
plt.show()
```
